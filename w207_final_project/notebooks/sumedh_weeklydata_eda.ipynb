```python
import pandas as pd
import numpy as np
import datetime
import os
import matplotlib.pyplot as plt
```

```python
main_df = pd.read_csv('../src/data/weekly_data.csv')
```

```python
print("Weekly dataset shape:",main_df.shape)
print(main_df.columns)
```

```python
main_df.head()
```

```python
main_df['ticker_symbol'].unique()
```

```python
print("Range of compound values:[{:.2f},{:.2f}]".format(main_df['compound'].min(),main_df['compound'].max()))
```

```python
# Plot compound values
plt.hist(main_df['compound'])
```

```python
# Plot log-transformed compound values
plt.hist(np.log(main_df['compound']))
```

```python
apple_df = main_df[main_df['ticker_symbol']=='AAPL']
```

```python
# Apple close value plot
apple_df = main_df[main_df['ticker_symbol']=='AAPL']
plt.scatter(apple_df['post_date'],apple_df['close_value'])
plt.title('Apple Stock Price Trend (2015-2020)')
plt.xlabel('Date')
plt.ylabel('Close Value ($)')
```

```python
# Apple compound plot
plt.figure(figsize=(10, 6))
plt.scatter(apple_df['post_date'],apple_df['compound'])
plt.title('Apple Compound Trend (2015-2020)')
plt.xlabel('Date')
plt.ylabel('Compound')
```
