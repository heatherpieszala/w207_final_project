```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
path = 'C:/Users/heath/Desktop/W207/data/'
main_df = pd.read_csv(path+'cleaned_data.csv')
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
main_df['sent_scores'] = main_df['body'].apply(lambda body: sid.polarity_scores(body))
```

```python
main_df
```

```python
#I wonder what would happen if we aggregated # reactions (comments, likes, retweets) per post and used that as a feature.
#My hypothesis is that it would have more influence
#also maybe positive and negative per ticker symbol per day has more influence? 
#The closing price and return per is the same across the stock
```

```python
main_df['compound']  = main_df['sent_scores'].apply(lambda score_dict: score_dict['compound'])
```

```python
main_df['tweet_sent'] = main_df['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
```

```python
main_df['reaction_num'] = main_df['comment_num'] + main_df['retweet_num'] + main_df['like_num']
```
